# Drop Un-needed Columns

```python
# Drop all but listed tables for outpatient
tennessee_outpatient_tableau = tennessee_outpatient_clean[[
    'provider_name',
    'apc',
    'apc_description',
    'beneficiaries',
    'comprehensive_apc_services',
    'average_estimated_total_submitted_charges',
    'average_medicare_allowed_amount',
    'average_medicare_payment_amount',
    'average_medicare_outlier_amount',
    'zip',
    'cbsa',
    'cbsa_title']]
tennessee_outpatient_tableau = tennessee_outpatient_tableau.dropna()
tennessee_outpatient_tableau
```

# Get Files Ready for Tableau

```python
tennessee_outpatient_tableau['zip']= tennessee_outpatient_tableau['zip'].astype(int)
tennessee_outpatient_tableau['apc']= tennessee_outpatient_tableau['apc'].astype(int)
tennessee_outpatient_tableau['cbsa']= tennessee_outpatient_tableau['cbsa'].astype(int)
tennessee_outpatient_tableau.info()
```
