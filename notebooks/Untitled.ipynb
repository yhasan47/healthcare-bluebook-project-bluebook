```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# import and set up dataframes
outpatient_clean = pd.read_csv('../data/outpatient_clean.csv', low_memory=False)
providers_clean = pd.read_csv('../data/providers_clean.csv', low_memory=False)
cbsa_to_name = pd.read_csv('../data/core-based-statistical-areas-cbsas-and-combined-statistical-areas-csas.csv', engine='python')

outpatient_clean.head(25)
```

```python
providers_clean.head(25)
```

```python
cbsa_to_name.info()
```

```python
cbsa_to_name.columns = cbsa_to_name.columns.str.replace(' ', '_').str.lower()
cbsa_to_name['cbsa_code'] = pd.to_numeric(cbsa_to_name['cbsa_code'], errors='coerce')
cbsa_to_name.info()
```

```python
outpatient_clean = pd.merge(left=outpatient_clean, right=cbsa_to_name, 
                            left_on='cbsa', right_on='cbsa_code', how='outer')
outpatient_clean.info()
```

```python
# Drop all but listed tables for outpatient
outpatient_tableau = outpatient_clean[[
    'provider_name',
    'apc',
    'apc_description',
    'beneficiaries',
    'comprehensive_apc_services',
    'average_estimated_total_submitted_charges',
    'average_medicare_allowed_amount',
    'average_medicare_payment_amount',
    'average_medicare_outlier_amount',
    'zip',
    'cbsa',
    'cbsa_title']]
outpatient_tableau = outpatient_tableau.dropna()
outpatient_tableau
```

```python
# Drop all but listed tables for proviers
providers_tableau = providers_clean[[
    'nppes_provider_last_org_name',
    'provider_type',
    'place_of_service',
    'hcpcs_code',
    'hcpcs_description',
    'line_srvc_cnt',
    'bene_unique_cnt',
    'average_medicare_allowed_amt',
    'average_submitted_chrg_amt',
    'average_medicare_payment_amt',
    'average_medicare_standard_amt']]
providers_tableau = providers_tableau.dropna()
providers_tableau.head(50)
```
